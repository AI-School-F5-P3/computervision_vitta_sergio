```python
# Instalar roboflow
!pip install roboflow

from roboflow import Roboflow
import cv2
import numpy as np

# Inicializar Roboflow y cargar tu modelo
rf = Roboflow(api_key="TU_API_KEY")
proyecto = rf.workspace("TU_WORKSPACE").project("NOMBRE_PROYECTO")
modelo = proyecto.version("NUMERO_VERSION").model

# Ejemplo de cómo hacer predicciones
# Con una imagen
prediccion = modelo.predict("ruta/a/tu/imagen.jpg", confidence=40, overlap=30).json()

# Con la webcam (ejemplo)
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break
        
    # Hacer predicción en el frame
    prediccion = modelo.predict(frame, confidence=40, overlap=30).json()
    
    # Dibujar las detecciones
    for prediccion in predicciones:
        x1 = prediccion['x'] - prediccion['width'] / 2
        y1 = prediccion['y'] - prediccion['height'] / 2
        x2 = prediccion['x'] + prediccion['width'] / 2
        y2 = prediccion['y'] + prediccion['height'] / 2
        
        cv2.rectangle(frame, 
                     (int(x1), int(y1)), 
                     (int(x2), int(y2)), 
                     (0, 255, 0), 
                     2)
        
        # Añadir etiqueta
        cv2.putText(frame, 
                    prediccion['class'], 
                    (int(x1), int(y1) - 10), 
                    cv2.FONT_HERSHEY_SIMPLEX, 
                    0.5, 
                    (0, 255, 0), 
                    2)
    
    cv2.imshow('Detecciones', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
